### Objetivos
Se pide a partir de la utilización de un divisor de tensión resistivo obtener la tensión de salida del mismo causando un error de método al empeorar artificialmente la resistencia voltimétrica del instrumento de sensor. Se pide obtener 2 errores de método diferentes:


1) Error de metodo despreciable

Antes que nada se busco una resistencia de Thevenin baja para que $ R_{V} $ tenga la influencia necesaria para despreciar el error de metodo. Al no conocer la resistencia de la fuente de tension, tuvimos en cuenta solo las que agregamos en el divisor resistivo. Por otra parte consultamos el manual del multimetro para conocer su resistencia interna.  

Valores: 
$$ R_{A} = 2K\Omega $$ 
$$ R_{B} = 2K\Omega $$ 
$$ V_{F} = 4 V $$
$$ R_{F} = N/D $$ 
$$ R_{V} = 10M\Omega $$

Tolerancia de las resistencias: 5% 

Calculamos la resistencia de Thevenin: 

$$ R_{th} = \frac{R_{A}R_{B}}{R_{A}+R_{B}} $$

$$ R_{th} = 1000\Omega $$

$$ e_{MR} = -\frac{R_{th}}{R_{th}+R_{V}} $$ 

Como $ R_{V} $ es muy grande con respecto a $ R_{th} $ el error de metodo se vuelve despreciable, como lo solicita el enunciado. 

$$ e_{MR} = -0.01\% $$ 

Tomamos valores de tension para realizar la comparacion entre el error de metodo teorico y el practico. En ambos casos el resultado fue un bajo porcentaje. 

Comparandolo con la tension: 
$$Vi_{1} = 1.973 $$
$$Vi_{2} = 1.973$$
$$Vi_{3} = 1.974 $$
$$Vi_{4} = 1.973 $$
$$Vi_{5} = 1.973$$

$$ V_{err} = -1\% $$ 

2) Error de metodo del 20% 

Para el segundo ensayo, introducimos una resistencia mucho mas baja en paralelo a la resistencia de entrada del multimetro para producir una $R_{V}$ y lograr el error de metodo solicitado. 

Resistencia de entrada del multimetro: $ R_{Vm} $ 
Resistencia que agregamos : $ R_{V1} $ 

Por eso calculamos, manteniendo el mismo valor para $R_{th}$: 

$$ e_{MR} = -\frac{R_{th}}{R_{th}+R_{V}} $$ 

$$ -0.2 = -\frac{1000}{1000+R_{V}} $$ 

$$ -0.2 = -\frac{1000}{1000+R_{V}} $$ 

$$ R_{V} = 4000 $$ 

Como $ R_{Vm} = 10M\Omega$ 

$$ R_{V1} \thickapprox 4K\Omega $$ 

Pero como la resistencia mas cercana de la que disponemos es $3.3K\Omega$, haremos los calculos para ese valor: 

Valores: 
$$ R_{A} = 2K\Omega $$ 
$$ R_{B} = 2K\Omega $$ 
$$ V_{F} = 4 V $$
$$ R_{F} = N/D $$ 
$$ R_{V1} = 3.3k\Omega $$
$$ R_{Vm} = 10M\Omega$$

Tolerancia de las resistencias: 5% 

Calculamos $R_{V}$: 

$$ R_{V} = 3298.911\Omega $$

$$ e_{MR} = -\frac{R_{th}}{R_{th}+R_{V}} $$ 

$$ e_{MR} = -23.25\% $$ 

Comparandolo con la tension: 
$$Vi_{1} = 1.524 $$
$$Vi_{2} = 1.525 $$
$$Vi_{3} = 1.524 $$
$$Vi_{4} = 1.523 $$
$$Vi_{5} = 1.524 $$

$$ V_{err} = -24\% $$ 

Como podemos observar los valores de tension se hacen muy distintos a los que deberiamos estar midiendo, lo que evidencia el error de metodo de mas del 20% que obtuvimos.
